```python
import requests
from bs4 import BeautifulSoup
```

```python
contest_url = 'http://hustle-sa.ru/forum/index.php?showtopic=4909'
html_text = requests.get(contest_url).text

soup = BeautifulSoup(html_text)
```

```python
posts = soup.select('td.post2')
len(posts)
```

```python
import re
pattern = re.compile("^Результаты турнира:")

def is_result_post(post):
    for b in post.select('b'):
        if pattern.match(b.text):
            return True
    return False
```

```python
result_post = [p for p in posts if is_result_post(p)][0]
```

```python
def all_children(root):
    return [child for child in root.children]
```

```python
results_div = result_post.find('div')
```

```python
def get_node_text(node):
    if hasattr(child, 'title'):
        return child.title()
    elif hasattr(child, 'text'):
        return child.text
    return ''

def get_page_raw_text_inner(root, strings):
    if not hasattr(root, 'children'):
        text = root.title()
        if text != ' ':
            strings.append(root.title().lower())
        return
    
    children = all_children(root)
    
    for child in children:
        get_page_raw_text_inner(child, strings)
        
def get_page_raw_text(root):
    strs = []
    get_page_raw_text_inner(root, strs)
    return strs
```

```python
for line in get_page_raw_text(soup):
    print(line)
```
