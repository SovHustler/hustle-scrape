### Загрузка данных

```python
raw_crosses = pd.read_csv(
    'dataset/crosses.csv', 
    names=["competitor_id", "competition_id", "judge_name", "category", "phase", "passed"],
    dtype={
        "competitor_id": str,
    }
).\
    dropna() # смотрим только на зарегистрированных спортсменов

raw_crosses.competitor_id = crosses.competitor_id.astype(int)

raw_crosses['syntethic_competitor_id'] = crosses.competitor_id - min(crosses.competitor_id)

popular_judges = [
    'ирина кудрявцева', 'кудрявцев михаил', 'лебедев сергей',
    'александр милованов', 'екатерина николаева', 'арутюнян евгений',
    'дутчак татьяна', 'катунин павел', 'андрей коньков', 
    'мавлатов юрий', 'дубровин игорь', 'анастасия лукьянова',
    'волконский иван', 'вероника лебедева'
]

raw_crosses['is_popular_judge'] = raw_crosses.judge_name.apply(lambda name: name in popular_judges)
```

```python
# выдадим спортсменам синтетический инкрементные идентификаторы

id_map = {}
i = 0
for id in raw_crosses.competitor_id:
    if id not in id_map:
        id_map[id] = i
        i += 1
    
raw_crosses['syntethic_competitor_id'] = [id_map[id] for id in raw_crosses.competitor_id]
```
